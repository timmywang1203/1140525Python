```python
#Cake
#Google 試算表連結代碼（ID）：1H0bLxlsiEEkOk06wAZl1BoxmOlfd5pVL
import pandas as pd
import pandas as pd
ID=input("請輸入已共享的 Google 試算表連結代碼（ID）：")
Filename= "https://docs.google.com/uc?id="+ID.strip()+"&export=download"
print(f"已產生下載網址：{Filename}")

Financial_Data= pd.read_excel(Filename,sheet_name="Financial_Data")
print(Financial_Data)
ESG_Score= pd.read_excel(Filename,sheet_name="ESG_Score")
print(ESG_Score)
OTHERS= pd.read_excel(Filename,sheet_name="OTHERS")
print(OTHERS)
Financial_Data= pd.read_excel(Filename,sheet_name="Financial_Data")
print(Financial_Data)
ESG_Score= pd.read_excel(Filename,sheet_name="ESG_Score")
print(ESG_Score)
OTHERS= pd.read_excel(Filename,sheet_name="OTHERS")
print(OTHERS)
```

# 第二題、合併資料集（用pd.merge 或 pd.concat）(5%)
#### 請將上面三個資料集（Financial_Data、ESG_Score、OTHERS）合併成一個資料集(Company_ESG)

```python
#Cake
import pandas as pd

Company_ESG = pd.merge(Financial_Data, ESG_Score, on='CompanyID', how='inner')
Company_ESG = pd.merge(Company_ESG, OTHERS, on='CompanyID', how='inner')

print(Company_ESG.head())
```

# 第七題、重新定義變數涵意及新增變數(18% = 3% + 10% + 5%)
#### (1)建立一個空的list變數(New_Group)
#### (2)利用for迴圈及ifelse判斷式將「資料集Company_ESG」中的ESG_Overall重新定義成3組別：
- 小於60分(不含60)
- 60-80分(不含80)
- 80分以上

#### (3)在「資料集Company_ESG」中新增一個欄位為「ESG成績分組」。

```python
#Cake
ESG_Overall_x=list()
for i in Company_ESG["ESG_Overall_x"]:
    if i<60:
        ESG_Overall_x.append("小於60分(不含60")
    elif i<80:
        ESG_Overall_x.append("60-80分(不含80)")
    else:
        ESG_Overall_x.append("80分以上")
ESG_Overall_x
```

```python
#Cake
Company_ESG["ESG成績分組"]=ESG_Overall_x

Company_ESG
```

# 第九題、計算統計量(10% = 4% + 3% + 3%)
#### 請以「資料集ESG_NumericData」計算：
1. 所有欄位個別的平均數、標準差
2. EnergyConsumption的中位數
3. 所有變數的變異係數矩陣

```python
#Cake
ESG_NumericData_2 = ESG_NumericData.dropna()
ESG_NumericData_2['ESG_Overall_x'].mean()
ESG_NumericData_2['ESG_Overall_x'].std()
print("\n所有欄位個別的平均數:")
print(ESG_NumericData_2.mean())

print("\n所有欄位個別的標準差:")
print(ESG_NumericData_2.std())

print("\nEnergyConsumption_x 的中位數:")
print(ESG_NumericData_2['EnergyConsumption_x'].median())

print("\n所有變數的變異係數矩陣:")
print(ESG_NumericData_2.std() / ESG_NumericData_2.mean())
```
